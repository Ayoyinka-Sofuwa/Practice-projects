```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hist2d(x, y, bins=20, cmap='Greens')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

```python
from scipy.stats import gaussian_kde
```

```python
np.meshgrid?
```

```python
# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape),
           origin='lower', aspect='auto',
           extent=[-3.5, 3.5, -6, 6],
           cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
```

### plt.subplots?

* Returns
-------
fig : `~.figure.Figure`

ax : `.axes.Axes` object or array of Axes objects.
    *ax* can be either a single `~matplotlib.axes.Axes` object or an
    array of Axes objects if more than one subplot was created

```python
plt.style.use('classic')
x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label= 'Sine')
ax.plot(x, np.cos(x), '--g', label = 'Cosine')
ax.axis('equal')

#here we tell the program to return 1000 numbers between 0&10 and
#make them equally spaced.
#then we tell the program to plot on the axes, the value of x and the 
#sine of x, also the cosine of x, using a straight, unbroken line for the sine
#and broken lines for the cosine.

#the ax.axis function/method is to scale the whole graph appropriately

leg = ax.legend(loc= 'lower right', frameon=False);
```
